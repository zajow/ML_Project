```python
import pandas as pd
df = pd.read_csv("Loan_default.csv")
df.head()
```

```python
# simple_random สุ่มค่าแบบ Random
loan_simple_random = df.sample(n=1000, random_state=42)
```

```python
# เปลี่ยนชื่อ Feature
df.rename(columns={
    'LoanID': 'Loan_ID',
    'LoanAmount': 'Loan_Amount',
    'Income': 'Income(USD)',
    'CreditScore': 'Credit_Score',
    'InterestRate': 'Interest_Rate(%)',
    'DTIratio': 'DebtToIncomeRatio',
    'LoanTerm': 'Loan_Term(months)',
}, inplace=True)
df.head()
```

```python

selected_columns = [
    "MonthsEmployed",
    "Education",
    "MaritalStatus",
    "HasDependents",
    "LoanPurpose",
    "HasCoSigner",
    "Loan_ID"
]

df = df.drop(columns=selected_columns)
df
```

```python
# One-hot Encoding (แก้คำเป็นตัวเลข)
#EmploymentType 
employment_dummies = pd.get_dummies(df['EmploymentType'], prefix='EmploymentType')
df = pd.concat([df.drop('EmploymentType', axis=1), employment_dummies], axis=1)
df.head()
```

```python
# แปลง HasMortagage เป็น 0,1
df['HasMortgage'] = df['HasMortgage'].replace({"Yes": True, "No": False})
df.head()
```

```python
%pip install -q scikit-learn

from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(
    df, test_size=0.1, stratify=df["Default"], random_state=123)

df_test
```

```python
df['Score_per_Amount'] = df['Credit_Score'] / df['Loan_Amount'].replace(0, pd.NA)

threshold = df['Score_per_Amount'].median()

df['IsRisked'] = (df['Score_per_Amount'] < threshold).astype(int)

print(f"Threshold (median Score_per_Amount): {threshold:.4f}")
print(df[['Credit_Score','Loan_Amount','Score_per_Amount','IsRisked']].head())

print(df['IsRisked'].value_counts())

df
```
